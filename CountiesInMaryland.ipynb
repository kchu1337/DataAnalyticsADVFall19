```python
import requests #library used to connect to a website
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#specify the url
URL = "https://en.wikipedia.org/wiki/List_of_counties_in_Maryland#List_of_counties"
```

```python
page = requests.get(URL)
soup = BeautifulSoup(page.text, "lxml")
```

```python
right_table=soup.find('table', class_='wikitable sortable')
print (right_table.prettify())
```

```python
header_list = []
iterator =0;
for row in right_table.findAll("th"):
    text =row.find(text=True)
    print(text)
    header_list.append(text)
    if '\n' in text:
        break;
```

```python
county_name_list = []
for row in right_table.findAll("th"):
    county_name = (row.find(text=True))
    if county_name in header_list:
        county_name
    else:
        county_name_list.append(county_name)
print(county_name_list)
```

```python
data=[[],[],[],[],[],[],[],[],[],[]]
for row in right_table.findAll("tr"):
    
    #get all the <td> tags for each <tr> tag
    cells = row.findAll('td')
    if len(cells)==10: 
        for i in range(len(cells)):
            data[i].append(cells[i].find(text=True))
            
data.insert(0, county_name_list)
print(data)
```

```python
df=pd.DataFrame()
for i in range(len(data)):
    header = header_list[i]
    df[f'{header}'] = data[i]
    print(header)
    print(data[i])
df.head()
```
